```python
#必要なライブラリをインポート
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
import numpy as np

#ファイルをデータフレームとして呼び込み、マージする。
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

#出現頻度をカウント
y = df['カテゴリー'].value_counts()
print(y)

#棒グラフ化する
x = np.array([1, 2, 3, 4, 5, 6])
label =["弁当", "菓子類", "飲料水", "デザート", "雑貨", "図書・新聞"]
plt.title("各カテゴリーの出現頻度")
plt.bar(x, y, tick_label=label)
plt.show()

#データフレームをカテゴリーごとにグループにまとめ、在庫と注文数の平均をみる。
df.groupby('カテゴリー')['在庫', '注文数'].mean()

```

```python
#データフレームをカテゴリーごとにグループにまとめ、在庫と注文数の基本統計量をみる。
df.groupby('カテゴリー')['在庫', '注文数'].describe()
```
